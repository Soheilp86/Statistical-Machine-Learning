### `if else` and `elif` statements

```python
## if, else statement!
if (x > 50) or (x < -50):
    print("|x| > 50")
else:
    print("|x| <= 50")
```

```python
## if else elif statement

if (x > 50) or (x < -50):
    print("|x| > 50")
elif x == 50:
    print("x = 50")
else:
    print("|x| < 50")
```

## Loops

###`For` Loops

A for loop runs a code a certain number of times. It starts with a `for statement`and then a comma and then indent and the code you want to repeat.


For reading more, here is the documentation https://docs.python.org/3/tutorial/controlflow.html#for-statements.

```python
for i in range(10): ## this will range from 0 to 9, i.e. 10 total numbers
    print(i)
```

```python
## You can specify a start and stop like so, again at intervals of 1
for i in range(2, 11):
    print(i)
```

```python
## you can also specify a spacing with a third "step" argument
## this goes from 71 to 142 in steps of 3
for i in range(-2, 8, 2):
    print(i)
```

```python
for names in ["Sara", "Soheyl", "Medi"]:
    print(names)
```

We can use a loop to create a list. For example, let's say we want to creat a list of even numbers from 0 to 100.

```python
##
even_numbers = [] ## start with am empty list
for i in range(101): ## for i in raage of 0 and 100,
    if i % 2 == 0: ## if the remainder of i devided by two is 0,
        even_numbers.append(i) ## add i to the list


print(even_numbers)
```

```python
## Alternative:
even_numbers = [i for i in range(101) if i % 2 == 0] ## this says collect i if i is between 0 and 100 and
                                                     ##also if when devided by two the remainder is zero
print(even_numbers)
```

```python
## Let's print out all numbers that are
## divisible by 7 or 11 between 1 and 100

## first the for loop
for i in range(1,101):
    ## our indent friend
    ## now the conditonal
    if i%7 == 0:
        ## oh boy, two indents!
        print(i,"is divisible by 7.")
    if i%11 == 0:
        print(i,"is divisible by 11.")
```

### `While` Loops

Another type of loop is a `while` loop. This runs the code until a certain logical condition is no longer valid.

For reading more, here is the documentation on while loops in python, https://docs.python.org/3/reference/compound_stmts.html#while.



```python
## First write while then the conditional statement and a colon
## then indent and write the code chunk you want executed

i = 10
print("Countdown time!")
while i > 0:
    print(i)
    i = i - 1

# Then return to your original indentation level
print("Lift off!")
```
