Let's start with a simple example:

write a function that takes in an integer and checks if that number is even:

```python
# Alternative Solution
def even_check(number):
    if number % 2 == 0:
        print(number, 'is even.')
    else:
        print(number, 'is odd.')
```

```python
even_check(4)
```

```python
# what happens if I plug in a float
even_check(5.2)
```

Let's change the above code so that if the input is not an integer prints a warning.

```python
def even_check(number):
    if type(number) == int:
      if number % 2 == 0:
        print(number, 'is even.')
      else:
        print(number, 'is odd.')
    else:
      print(' Warning: ', number, 'is not an integer')
```

```python
even_check(5.2)
```

Sometimes we need a function of several variables but also a default value for a variable.

```python
def divisible_by(x, y=2):
    if x%y == 0:
        print("Yes.", x, "is divisible by", y)
    else:
        print("No.", x, "is not divisble by", y)
```

```python
divisible_by(10)
```

```python
## we can change the default value:

divisible_by(10,3)
```

```python
## Write a function that takes in three numbers and return their avaerage.
def average(num1, num2, num3):
  return (num1 + num2 + num3)/3
```

```python
### Write a function that takes in a positive integer n and computes the sum of all numbers from 1 to n
def sum_numbers(n):
  sum = 0
  for i in range(1,n+1):
    sum += i
  return sum
```

```python
#This function computes how long it takes to pay off a loan. Modify it to also calculate your interest
import numpy as np

def pay_off(balance, annual_rate, monthly_payment):
  month = 0
  monthly_rate = np.exp(np.log(1 + annual_rate)/12.0) - 1
  monthly_factor = 1 + monthly_rate
  while balance >0:
    print("%4d: %4.2f" %(month, balance))
    balance = monthly_factor*balance - monthly_payment
    month = month + 1
  print("\n%d months to pay off your loan" %(month))
```

```python
pay_off(5000,0.09, 150 )
```
