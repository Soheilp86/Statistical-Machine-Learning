#Jupyter Notebook

Jupyter Notebook is an open-source web application that allows you to create and share documents containing live code (most commonly Python), equations, visualizations, and narrative text. It's widely used in data science, scientific research, and education because it supports interactive computing and reproducible research.  

We can run Jupyter Notebook using Anaconda, Google Colab, or other platforms that support it, making it versatile and accessible for various users. In this course we will be using Google Colab

##Jupyter Notebook using Google Colab

Google Colab provides an easy and powerful way to use Jupyter Notebooks without any local setup, and it's especially useful for projects requiring significant computational power. Additionally, you can save and access your notebooks directly from Google Drive, ensuring your work is securely stored and easily accessible.

### Step 1: Sign In to Google Colab

1. **Google Account:**
   - You need a Gmail account to use Google Colab. If you don’t have one, you can create it [here](https://accounts.google.com/SignUp).

2. **Access Google Colab:**
   - Go to [Google Colab](https://colab.research.google.com/) and sign in with your Gmail account.

3. **Create A New Notebook:**
   - Click on the “File” menu in the top-left corner and select “New notebook.” This will open a new Jupyter notebook in the browser.

## Jupyter Notebook using Anaconda

Anaconda is a popular open-source distribution of Python and R for data science and machine learning. It simplifies package management and deployment with the Conda package manager, making it easy to install and manage libraries and dependencies. Anaconda includes Anaconda Navigator, a user-friendly GUI to launch applications like Jupyter Notebook and Spyder without using the command line. It’s widely used for its ease of use, comprehensive library support, and robust environment management.



### Step 1: Install Anaconda

1. **Download Anaconda:**
   - Visit the [Anaconda website](https://www.anaconda.com/products/distribution) and download the installer for your operating system (Windows, macOS, or Linux).

2. **Install Anaconda:**
   - Follow the installation instructions for your operating system.

### Step 2: Launch Jupyter Notebook

1. **Open Anaconda Navigator:**
   - On Windows, open the Start Menu and search for "Anaconda Navigator."
   - On macOS or Linux, open the terminal and type `anaconda-navigator` and press Enter.

2. **Launch Jupyter Notebook:**
   - In Anaconda Navigator, click the "Launch" button under the Jupyter Notebook section. This will open a new tab in your default web browser with the Jupyter Notebook interface.

### Step 3: Create a New Notebook

1. **Select the Working Directory:**
   - Navigate to the folder where you want to save your notebook. You can do this by clicking on the folder icons.

2. **Create a New Notebook:**
   - Click on the "New" button on the right side of the screen and select "Python 3" (or any other kernel you prefer) from the dropdown menu. This will open a new notebook.

## My First Jupyter Notebook (JNB)

As I mentioned before, a JNB allows us to combine explanatory text (in Markdown blocks) with code (in Code blocks).

### Markdown Cells or Text cells

You can add explanatory text using Markdown cells. To create a Markdown cell, simply click on `+ Text` (top-right of this page). Write your text using Markdown syntax and run the cell with Shift + Enter. If you are familiar with latex, you can also write math formulas such as

$$x^2 +y^2 = 1$$

$$\sum^{n}_{i=1} x_i$$

$$∫ \sin(x) dx$$
